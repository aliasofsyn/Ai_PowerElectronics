### pytorch

```python
# Test PyTorch installation and create a simple neural network example
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Check if CUDA is available
print(f"CUDA available: {torch.cuda.is_available()}")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1) * 0.5

# Convert to PyTorch tensors
X_tensor = torch.FloatTensor(X).to(device)
y_tensor = torch.FloatTensor(y).to(device)

# Define a simple neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(1, 10)
        self.layer2 = nn.Linear(10, 5)
        self.layer3 = nn.Linear(5, 1)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.layer3(x)
        return x

# Initialize model, loss function and optimizer
model = SimpleNN().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training loop
epochs = 200
losses = []

for epoch in range(epochs):
    # Forward pass
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    losses.append(loss.item())
    
    if (epoch + 1) % 50 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Visualization
plt.figure(figsize=(15, 5))

# Plot 1: Training Data and Predictions
plt.subplot(1, 2, 1)
with torch.no_grad():
    predicted = model(X_tensor).cpu().numpy()
plt.scatter(X, y, color='blue', alpha=0.5, label='Original Data')
plt.scatter(X, predicted, color='red', alpha=0.5, label='Predictions')
plt.xlabel('Input (X)')
plt.ylabel('Output (y)')
plt.title('Training Data vs Predictions')
plt.legend()

# Plot 2: Loss over epochs
plt.subplot(1, 2, 2)
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Time')

plt.tight_layout()
plt.show()
```

### pyswarms

```python
# Test PySwarms installation with visualization
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
from pyswarms.utils.plotters import plot_contour, plot_surface
import matplotlib.pyplot as plt
import numpy as np

# Set up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}

# Initialize swarm with 30 particles in 2 dimensions
optimizer = ps.single.GlobalBestPSO(n_particles=5, dimensions=2, options=options)

# Create figure for visualization
plt.figure(figsize=(15, 10))

# Store history for plotting
pos_history = []

# Create a custom objective function that records positions
def objective_function(positions):
    # Calculate the actual Sphere function value
    cost = fx.sphere(positions)
    # Record the current positions (only when called by the optimizer)
    if positions.shape == optimizer.swarm.position.shape:
        pos_history.append(optimizer.swarm.position.copy())
    return cost

# Optimize with our custom objective function
cost, pos = optimizer.optimize(objective_function, iters=50)

# Create plots for iterations 10, 50, and 80
iterations = [1, 20, 40]
for idx, iter_num in enumerate(iterations, 1):
    plt.subplot(2, 2, idx)
    
    # Plot contour of Sphere function
    x_grid = np.linspace(-5, 5, 100)  # Sphere function typical bounds
    y_grid = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x_grid, y_grid)
    Z = np.zeros_like(X)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i,j] = fx.sphere(np.array([[X[i,j], Y[i,j]]]))[0]
    
    plt.contour(X, Y, Z, levels=20)
    
    # Plot particle positions at the specific iteration
    plt.scatter(pos_history[iter_num][:, 0], 
                pos_history[iter_num][:, 1], 
                c='red', 
                marker='o', 
                alpha=0.5,
                label=f'Particle positions at iteration {iter_num}')
    
    plt.title(f'Particle positions at iteration {iter_num} (Sphere)')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.legend()

# Add final results text in the fourth subplot
plt.subplot(2, 2, 4)
plt.text(0.1, 0.5, 
         f"Optimization Results:\n\nBest cost achieved: {cost:.4f}\nBest position: [{pos[0]:.4f}, {pos[1]:.4f}]",
         fontsize=12)
plt.axis('off')

plt.tight_layout()
plt.show()

print("PySwarms optimization of Sphere function complete!")
print(f"Best cost achieved: {cost}")
print(f"Best position found: {pos}")
```

### pygad

```python
# Import the required libraries
import pygad
import numpy as np
import matplotlib.pyplot as plt

# Define a simple fitness function to test PyGAD
# Updated to accept 3 parameters as required by PyGAD 2.20.0
def fitness_func(ga_instance, solution, solution_idx):
    # Simple sphere function (minimize)
    return -np.sum(solution**2)  # Negative because PyGAD maximizes by default

# Define the parameters for the genetic algorithm
num_generations = 50
num_parents_mating = 4
sol_per_pop = 8
num_genes = 2

# Define the range for the genes
gene_space = {'low': -5, 'high': 5}

# Define callback function to save population at each generation
populations_history = []

def on_generation(ga_instance):
    populations_history.append(ga_instance.population.copy())

# Create an instance of the GA class
ga_instance = pygad.GA(num_generations=num_generations,
                      num_parents_mating=num_parents_mating,
                      sol_per_pop=sol_per_pop,
                      num_genes=num_genes,
                      gene_space=gene_space,
                      fitness_func=fitness_func,
                      on_generation=on_generation)

# Run the genetic algorithm
ga_instance.run()

# Get the best solution
solution, solution_fitness, solution_idx = ga_instance.best_solution()

# Print the results
print("PyGAD is installed and running correctly!")
print(f"Best solution found: {solution}")
print(f"Fitness value of the best solution: {solution_fitness}")
print(f"Generation of the best solution: {ga_instance.best_solution_generation}")

# Create a figure to visualize both fitness and population positions
plt.figure(figsize=(15, 10))

# Plot the fitness value across the generations in a separate subplot
plt.subplot(2, 2, 1)
plt.plot(ga_instance.best_solutions_fitness)
plt.title('Fitness Value Over Generations')
plt.xlabel('Generation')
plt.ylabel('Fitness Value')
plt.grid(True)

# Select iterations to visualize (beginning, 1/3, 2/3, and end)
iterations_to_plot = [0, num_generations//3, 2*num_generations//3, num_generations-1]

# Create a sphere function for contour plotting
def sphere(x, y):
    return x**2 + y**2

# Create grid for contour plot
x_grid = np.linspace(-5, 5, 100)
y_grid = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_grid, y_grid)
Z = sphere(X, Y)

# Plot positions at selected iterations
for i, iter_num in enumerate(iterations_to_plot[1:], start=1):  # Start from the second iteration
    plt.subplot(2, 2, i+1)
    
    # Plot contour of the sphere function
    plt.contour(X, Y, Z, levels=20)
    
    # Plot individual positions
    population = populations_history[iter_num]
    plt.scatter(population[:, 0], population[:, 1], 
                c='red', marker='o', alpha=0.7,
                label=f'Individuals at generation {iter_num}')
    
    # Mark the best individual
    best_idx = np.argmax([fitness_func(ga_instance, ind, idx) for idx, ind in enumerate(population)])
    plt.scatter(population[best_idx, 0], population[best_idx, 1], 
                c='blue', marker='*', s=200, alpha=1.0,
                label='Best individual')
    
    plt.title(f'Population at generation {iter_num}')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.legend()
    plt.grid(True)

# Add final results text in the fourth subplot
plt.subplot(2, 2, 4)
plt.text(0.1, 0.5, 
         f"Optimization Results:\n\nBest fitness achieved: {solution_fitness:.4f}\nBest position: [{solution[0]:.4f}, {solution[1]:.4f}]",
         fontsize=12)
plt.axis('off')

plt.tight_layout()
plt.show()
```

### pymoo

```python
# Import pymoo and verify installation
import numpy as np
from pymoo.core.problem import Problem
from pymoo.algorithms.moo.nsga2 import NSGA2
# Updated imports for newer pymoo versions
from pymoo.operators.sampling.rnd import FloatRandomSampling
from pymoo.operators.crossover.sbx import SBX
from pymoo.operators.mutation.pm import PM
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter

# Define a simple test problem (ZDT1)
class ZDT1(Problem):
    def __init__(self):
        super().__init__(n_var=30,
                         n_obj=2,
                         n_constr=0,
                         xl=np.zeros(30),
                         xu=np.ones(30))

    def _evaluate(self, x, out, *args, **kwargs):
        f1 = x[:, 0]
        g = 1 + 9.0 / (self.n_var - 1) * np.sum(x[:, 1:], axis=1)
        f2 = g * (1 - np.sqrt(f1 / g))
        out["F"] = np.column_stack([f1, f2])

# Setup and run a simple optimization
problem = ZDT1()
algorithm = NSGA2(
    pop_size=100,
    n_offsprings=100,
    sampling=FloatRandomSampling(),
    crossover=SBX(prob=0.9, eta=15),
    mutation=PM(eta=20),
    eliminate_duplicates=True
)

# Run the algorithm for a few generations
res = minimize(problem,
               algorithm,
               ('n_gen', 10),
               seed=1,
               verbose=True)

# Plot the Pareto front
plot = Scatter()
plot.add(res.F, color="red", label="Pareto front")
plot.show()

print(f"Number of solutions on the Pareto front: {len(res.F)}")
print("Pymoo installation is working correctly!")
```
